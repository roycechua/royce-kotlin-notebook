```kotlin
// 1 - Printing data on the console
println("Hello") // semi-colons not needed
print("World")
print(", here I am"); // even if you do they'll still work
```

```kotlin
// 2 - Variables
val fixed = 12 // constant value (immutable) and fixed data type Int
println(fixed) // types are inferred (type inference) based on the value
println(fixed is Int)
```

```kotlin
fixed = 13
```

```kotlin
var changing = 1 // mutable variable value but fixed data type
println(changing)
changing = 2
println(changing)
```

```kotlin
changing = "a"
```

```kotlin
var something = null
println(something)
```

```kotlin
var uninitialized: String
println(uninitialized)
uninitialized = "a"
```

```kotlin
val price = 5.0
val quantity = 3
// val cost: Int = price * quantity // cost is incorrect resulting type
// val cost: Double = price * quantity // double
val cost = price * quantity
println(cost)
```

```kotlin
val word = "hello "
// val words = word * 3 (doesn't exist here)
val words = word.repeat(3) // string method
println(words)
```

```kotlin
val message1 = "hello"
val message2 = "world"
println(message1 + " " + message2)
```

```kotlin
// 3 - Printing variables with strings
val price = 5.0
println("The price of this item is ${price}")
println("The price of this item is $price")
```

```kotlin
val message = "I wanted to say"
val word = "hello"
println("$message ${word.repeat(3)}") // method calls need the curly braces {}
```
