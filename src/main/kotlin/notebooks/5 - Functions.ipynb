# 5 - Functions

https://kotlinlang.org/docs/basic-syntax.html#functions

Default return type is Unit or Void in other languages

General syntax
``` kotlin
fun functionName(param: type): returnType {
  // any codes
  // return // if not Unit
}
```

````kotlin
fun greet() {
    println("Hello World")
}
````

````kotlin
greet()
````

````kotlin
fun sum(a: Int, b: Int): Int {
    return a + b
}
````

````kotlin
println(sum(1,2))
````

optional parameters

````kotlin
fun greetV2(message: String = "Hi there!"): String {
    return message
}

println(greetV2())
println(greetV2("Yow"))
````

function as values (type inferred) - anonymous functions

````kotlin
val getRectangleArea = fun (x: Int, y: Int) = x * y
println(getRectangleArea(3,5))
````

````kotlin
val PI = 3.14

val getCircleArea = fun (r: Int): Double {
    return PI * (r * r)
}

println(getCircleArea(3))
````
