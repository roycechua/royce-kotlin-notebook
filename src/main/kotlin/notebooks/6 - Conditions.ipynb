# If statement
As block (common) and as values (in Javascript/Typescript) these are called ternary expressions

```kotlin
val condition = true // boolean

if (true) {
    println("Do this")
} else {
    println("Instead")
}
```

```kotlin
println("Enter your age: ")
val age = readln().toInt()
val canVote = if (age >= 18) "yes" else "No"

println("Of legal voting age: $canVote")
```

The symbol ?: takes the right-hand value if the left-hand value is null (the elvis operator).

```kotlin
val status = null
val message = status ?: "Default Message"
println("The message is $message")
```

# When Statement

Functions similarly to switch case but more flexible

```kotlin
val statusCode = 400
val response = when (statusCode) {
    200 -> "OK!"
    400 -> "Bad Request"
    else -> "Internal Server Error"
}

println("Server response: $response")
```
