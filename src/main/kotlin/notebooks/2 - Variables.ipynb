# 2 - Variables

Using `val` is equivalent to using a `const` in Javascript/Typescript where the value is immutable.

Using `var` is equivalent to using `let` in Javascript/Typescript where the value can

The data type of the variable is either
1. inferred based on the value (default) or
2. the type is explicitily specified
but the variable's data type cannot be changed

The only exception where the variable can have 2 data types if the data type is marked as optional or the value could either be a value of the specified data type or null

```kotlin
val fixed = 12 // constant value (immutable) and fixed data type Int
println(fixed) // types are inferred (type inference) based on the value
println(fixed is Int)
```

```kotlin
fixed = 13
```

```kotlin
var changing: Int = 1 // mutable variable value but fixed data type
println(changing)
changing = 2
println(changing)
```

```kotlin
changing = "a"
```

```kotlin
// If a data type is not explicitly marked as optional using ?, it's data type cannot be changed from its initially defined data type ex. Nothing?
var something = null
println(something)
```

```kotlin
something = 3
```

```kotlin
// the value of variable another can now be either Int or Null
var another: Int? = null
another = 3
println(another)
```

```kotlin
var uninitialized: String
println(uninitialized)
uninitialized = "a"
```

Better to initialize a variable with a default value unless necessary to leave uninitialized
