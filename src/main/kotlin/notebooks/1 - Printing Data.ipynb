Print variables or data with string templates `$` and `${}`. Expressions can be evaluated inside `${1+1}`

```kotlin
val price = 5.0
println("The price of this item is $${5}")
println("The price of this item is $${5+10}")
println("The price of this item is $${price}")
println("The price of this item is $$price")
```

```kotlin
val message = "I wanted to say"
val word = "hello"
println("$message ${word.repeat(3)}") // method calls need the curly braces {}
```
