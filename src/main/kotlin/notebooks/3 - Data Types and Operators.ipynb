# 3 - Data Types and Operators

# Int Data Type

operators (+, -, *, /, %)

```kotlin
val groupAParticipants = 20
val groupBParticipants = 30

val totalSnacks = 10
val attendees = 7

val price = 5
val quantity = 3

val freebies = 12
val groups = 3 // how many members should be in each group for everyone to receive a freebie in their group

val value = 121
```

Variables can evaluate expressions

```kotlin
val totalParticipants = groupAParticipants + groupBParticipants
println("Total participants: $totalParticipants")
```

```kotlin
val totalExcessSnacks = totalSnacks - attendees
println("Total excess snacks: $totalExcessSnacks")
```

```kotlin
// Multiplication (*)
val cost = price * quantity
println("Total item cost: $$cost")
```

```kotlin
val freebiesPerGroup = freebies / groups
println("Freebie item per group: ${freebiesPerGroup} pcs")
```

```kotlin
val remainder = 121 % 2
println("The remainder is $remainder")
```

# Boolean Data Type

operators &&, ||, >, >=, <=, <

```kotlin
val legalVotingAge = true
val naturalCitizen = false
val canVote = legalVotingAge && naturalCitizen
println("Can vote: $canVote")
```

```kotlin
val isFruit = false
val isCandy = true
val canEat = isFruit || isCandy
println("Can eat $canEat")
```

```kotlin
val bigger = 3 > 2
val smaller = 3 < 5
println(bigger)
println(smaller)
```
